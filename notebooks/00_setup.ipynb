## Cell 00-03 | Install Base Dependencies

```python
import subprocess, sys

def pip(packages):
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q'] + packages)

print('Installing core ML packages...')
pip(['torch', 'torchvision', 'torchaudio'])
print('Installing computer vision packages...')
pip(['opencv-python-headless', 'Pillow', 'albumentations', 'timm', 'einops', 'kornia'])
print('Installing training utilities...')
pip(['wandb', 'tensorboard', 'tqdm', 'matplotlib', 'numpy', 'scipy', 'scikit-learn', 'pandas'])
print('Installing model-specific packages...')
pip(['ultralytics', 'transformers', 'datasets', 'accelerate', 'peft',
     'bitsandbytes', 'onnx', 'onnxruntime', 'huggingface_hub'])
print('All packages installed successfully')
```
